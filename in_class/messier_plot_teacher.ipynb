```python
from astroquery.vizier import Vizier
from astroquery.simbad import Simbad
import astropy.units as u
import astropy.coordinates as coord
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# use astroquery to get all of the messier objects from bibcode 1850CDT..1784..227M via simbad

messier_objects = []
for i in range(1,102):
    messier_objects.append("M{}".format(i))

simbad_table = Simbad.query_objects(messier_objects)
```

```python
simbad_table
```

```python
ra, dec = simbad_table["RA"], simbad_table["DEC"]

coords = coord.SkyCoord(ra, dec, unit=(u.hourangle, u.deg))
```

```python
# mollewiede projection
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(coords.ra.wrap_at(180 * u.deg).radian, coords.dec.radian)

# add equatorial grid
ax.set_xticklabels(["14h", "16h", "18h", "20h", "22h", "0h", "2h", "4h", "6h", "8h", "10h"])
ax.set_xlabel("Right Ascension")
ax.set_ylabel("Declination")
ax.grid(True)
```
